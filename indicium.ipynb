# Verificando outliers.

```python
# perceptivel que tem alguns valores bem discrepantes, vou fazer uma representao grafica para melhor vizualizao
dataframe.groupby(['price']).size()
```

```python
dataframe.boxplot(column='price', grid=False,
                           color=dict(boxes='blue', whiskers='black', medians='red'))
plt.title('Preo ')
plt.suptitle('')
plt.xticks(rotation=90)
plt.show()
# perceptivel que existem alguns outliers que podem interferir na analise do dataframe
```

```python
#aqui utilizo o medo IQR para definir os limites para os outliers
Q1 = dataframe['price'].quantile(0.25)
Q3 = dataframe['price'].quantile(0.75)
IQR = Q3 - Q1
limite_minimo = Q1 - 1.5 * IQR
limite_maximo = Q3 + 1.5 * IQR

#fao o tratamento para que ele no atija o limite maximo definido
dataframe_sem_outliers_preo = dataframe[
    (dataframe['price'] >= limite_minimo) & (dataframe['price'] <= limite_maximo)]

print(f"Original: {len(dataframe)}")
print(f"Sem outliers: {len(dataframe_sem_outliers_preo)}")
```

```python
#agora os dados esto mais consistentes
dataframe_sem_outliers_preo.groupby(['price']).size()
```

# Comparando colunas para ver padro e correlao.

```python
#crio um boxplot comparando a coluna bairro_group com price
sns.barplot(x=dataframe_sem_outliers_preo["bairro_group"], y=dataframe_sem_outliers_preo["price"], ci=None, palette="viridis")
plt.title("Preo Mdio por Tipo de Imvel")
plt.xlabel("Bairro")
plt.ylabel("Numero de reviews")
plt.xticks(rotation=45)
plt.show()
```

```python
# comparo a coluna bairro_group com numer_de_reviews
sns.barplot(x=dataframe_sem_outliers_preo["bairro_group"], y=dataframe_sem_outliers_preo["numero_de_reviews"], ci=None, palette="viridis")
plt.title("Preo Mdio por Tipo de Imvel")
plt.xlabel("Bairro")
plt.ylabel("Numero de reviews")
plt.xticks(rotation=45)
plt.show()
```

```python
#comparao da coluna bairro_group com a disponibilidade
sns.barplot(x=dataframe_sem_outliers_preo["bairro_group"], y=dataframe_sem_outliers_preo["disponibilidade_365"], errorbar=None, palette="viridis")
plt.title("Preo Mdio por Tipo de Imvel")
plt.xlabel("Bairro")
plt.ylabel("Numero de reviews")
plt.xticks(rotation=45)
plt.show()
```

```python
#comparao da coluna bairro_group com a coluna price

sns.barplot(x=dataframe_sem_outliers_preo["bairro_group"], y=dataframe_sem_outliers_preo["minimo_noites"], errorbar=None, palette="viridis")
plt.title("Preo Mdio por Tipo de Imvel")
plt.xlabel("Bairro")
plt.ylabel("minimo_noites")
plt.xticks(rotation=45)
plt.show()
```
